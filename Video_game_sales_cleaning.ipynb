# Cleaning

```python
from pyspark import SparkFiles
```

```python
# import dependencies
import pandas as pd
import numpy as np

# reads sales data
vgsales_df = pd.read_csv('Resources/vgsales-12-4-2019-short.csv')
vgsales_df.head(10)
```

```python
# read country data
country_df = pd.read_csv('Resources/video-games-developers.csv')
country_df.head(10)
```

```python
# merge two dataset
df = pd.merge(vgsales_df, country_df[['Developer', 'Country']], left_on='Publisher', right_on='Developer', how='left')
df.head(10)
```

```python
df.count()
```

```python
df.columns
```

```python
# Drop unnecessary columns 
clean_df = df.drop(['Developer_y','NA_Sales','PAL_Sales','JP_Sales','Other_Sales'], axis=1)
```

```python
clean_df['Total_Sales'] = clean_df.Total_Shipped.fillna(0) + clean_df.Global_Sales.fillna(0)
clean_df = clean_df.drop(['Total_Shipped','Global_Sales',], axis=1)
clean_df
```

```python
# Find null value
for column in clean_df.columns:
    print(f'Column {column} has {clean_df[column].isnull().sum()} null values')
```

```python
# Get the count of Zeros in column 'Total_Sales' 
count = (clean_df['Total_Sales'] == 0).sum()
print('Count of zeros in Total_Sales : ', count)
```

```python
# Drop unnecessary columns 
cleaned_data = clean_df.drop(['ESRB_Rating','Critic_Score','User_Score','Country'], axis=1)
cleaned_data
```

```python
# Find null value
for column in cleaned_data.columns:
    print(f'Column {column} has {cleaned_data[column].isnull().sum()} null values')
```

```python
cleaned_data_df = cleaned_data.dropna(subset=['Developer_x','Year'])
cleaned_data_df
```

```python
# Find null value
for column in cleaned_data_df.columns:
    print(f'Column {column} has {cleaned_data_df[column].isnull().sum()} null values')
```

```python
#drop the row with $0.00 in Total_Sales
cleaned_df = cleaned_data_df[cleaned_data_df['Total_Sales'] !=0]
cleaned_df
```

```python
# Get the count of Zeros in column 'Total_Sales' 
count = (cleaned_df['Total_Sales'] == 0).sum()
print('Count of zeros in Total_Sales : ', count)
```
