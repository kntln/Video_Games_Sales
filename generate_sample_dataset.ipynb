# Generate random data

```python
import pandas as pd
import random
import numpy as np
```

## Year vs Sales

```python
int_list = random.sample(range(10, 1000), 30)
sales = [x/10 for x in int_list]
print(sales)
```

```python
data = {'Year_Released': random.sample(range(1970, 2020), 30),
        'Total_Sales': sales
        }
df = pd.DataFrame(data, columns= ['Year_Released', 'Total_Sales'])
df
```

```python
df.to_csv('Sample_Datasets/sample_years_sales.csv', index = False, header=True)
```

## Critic score vs Sales

```python
score = [x/10 for x in random.sample(range(10, 100), 30)]
sales2 = [x/10 for x in random.sample(range(10, 1000), 30)]
critic_data = {'Critic_Score': score ,
        'Total_Sales': sales2
        }
critic_df = pd.DataFrame(critic_data, columns= ['Critic_Score', 'Total_Sales'])
critic_df
```

```python
critic_df.to_csv('Sample_Datasets/sample_critic_sales.csv', index = False, header=True)
```

## Various Variables vs Sales

```python
# Generate random number to mimic the encoding
genre = np.random.choice(20, 30, replace=True)
ESRB = np.random.choice(6, 30, replace=True)
platform = np.random.choice(40, 30, replace=True)
country = np.random.choice(20, 30, replace=True)
year_num = np.random.choice(45, 30, replace=True)

sales3 = [x/10 for x in random.sample(range(10, 1000), 30)]
userscore = [x/10 for x in random.sample(range(10, 100), 30)]
```

```python
encoded_data = {
    'Critic_Score': score,
    'User_Score': userscore,
    'Genre': genre,
    'ESRB_Rating': ESRB,
    'Platform': platform,
    'Year_released': year_num,
    'Country': country,
    'Total_Sales': sales3
}

encoded_df = pd.DataFrame(encoded_data, columns= [
    'Critic_Score',
    'User_Score',
    'Genre',
    'ESRB_Rating',
    'Platform',
    'Year_released',
    'Country',
    'Total_Sales'
])
encoded_df
```

```python
encoded_df.to_csv('Sample_Datasets/sample_encoded_data.csv', index = False, header=True)
```
